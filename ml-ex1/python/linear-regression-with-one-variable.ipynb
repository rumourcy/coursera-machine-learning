#### 假设空间

$h_\theta=\theta^Tx=\theta_0+\theta_1x_1$

```python
def h(X, theta):
    return np.dot(X, theta)
```

#### 损失函数

$J(\theta) = \frac{1}{2m}\sum_{i=1}^{m}{(h_\theta(x^{(i)})-y^{(i)})}^2$

```python
def J(X, y, theta):
    m = 2 * len(y)
    loss = h(X, theta) - y
    return np.dot(loss.T, loss) / m
```

```python
y = data[:, 1]
X = np.append(np.ones((len(y), 1)), data[:, 0].reshape((-1, 1)), axis=1)
theta = np.zeros(2)
alpha = 0.01
num_iters = 1500
```

```python
J(X, y, theta)
```

#### 梯度下降

$\theta_j := \theta_j - \alpha\frac{1}{m}\sum_{i=1}^{m}(h_\theta(x^{(i)}-y^{(i)})x_j^{(i)}$

```python
def gd(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros(num_iters)
    for i in range(num_iters):
        loss = h(X, theta) - y
        theta = theta - alpha * np.dot(X.T, loss) / m
        
        J_history[i] = J(X, y, theta)
    return theta, J_history
```

```python
theta, J_history= gd(X, y, theta, alpha, num_iters)
```

```python
plt.plot(np.arange(num_iters), J_history)
plt.xlabel('num of iters')
plt.ylabel('loss function')
plt.show()
```

```python
theta0_vals = np.linspace(-10, 10, 100)
theta1_vals = np.linspace(-1, 4, 100)

J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

for i in np.arange(len(theta0_vals)):
    for j in np.arange(len(theta1_vals)):
        theta = np.array([theta0_vals[i], theta1_vals[j]])
        J_vals[i, j] = J(X, y, theta)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(theta0_vals, theta1_vals, J_vals) 
```
